```
import random
from qiskit import QuantumCircuit
from qiskit_aer import Aer # Corrected import for Aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
n = 10 # Number of bits
alice_bits = [random.randint(0,1) for _ in range(n)]
alice_bases = [random.choice(['Z','X']) for _ in range(n)]
bob_bases = [random.choice(['Z','X']) for _ in range(n)]
bob_results = []
simulator = Aer.get_backend('qasm_simulator')
for bit,a_basis,b_basis in zip(alice_bits, alice_bases, bob_bases):
  qc = QuantumCircuit(1,1)
  if bit == 1:
    qc.x(0)
  if a_basis == 'X':
    qc.h(0)
  if b_basis == 'X':
    qc.h(0)
    qc.measure(0,0)
    # Corrected: Use simulator.run() instead of the deprecated execute()
    result = simulator.run(qc, shots=1).result()
    counts = result.get_counts()
    bob_results.append(int(list(counts.keys())[0])) # Corrected typo from apend to append

    # Key generation(only  where bases matching)
    key = [b for a,b,basis_match in zip(alice_bits, bob_results,
                                       [ab==bb for ab,bb in zip(alice_bases,bob_bases)])if
          basis_match]
print("Alice bits:", alice_bits)
print("Alice bases:", alice_bases)
print("Bob bases:", bob_bases)
print("Bob results:",bob_results)
print("Shared secret key(where bases match):", key)
```
